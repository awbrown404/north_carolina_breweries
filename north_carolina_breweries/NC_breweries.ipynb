```python
import requests
import pandas as pd
import json
from config import api_key
```

```python
# http://api.brewerydb.com/v2/{endpoint}/?key=abcdef
# http://api.brewerydb.com/v2/locations?key=KEY&locality=denver&region=colorado
base_url = "http://api.brewerydb.com/v2/"
endpoint = "locations/"
parameter = "&region=NC&p=1"

response_pg1 = requests.get(base_url + endpoint + "?key=" + api_key + parameter).json()
```

```python
print(json.dumps(response_pg1, indent=4, sort_keys=True))
```

```python
pg1_data = response_pg1["data"]
pg1_data
```

```python
breweries = []
city = []
state = []
latitude = []
longitude = []
brewery_ID = []
brewery_type = []

for result in pg1_data:
    breweries.append(result["brewery"]["name"])
    city.append(result["locality"])
    state.append(result["region"])
    latitude.append(result["latitude"])
    longitude.append(result["longitude"])
    brewery_ID.append(result["breweryId"])
    brewery_type.append(result['locationTypeDisplay'])
```
