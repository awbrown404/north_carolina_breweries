```python
# set up API call
# http://api.brewerydb.com/v2/{endpoint}/?key=abcdef
# http://api.brewerydb.com/v2/beers/

# url = (f'http://api.brewerydb.com/v2/brewery/{id}/?key={api_key}&withBreweries=Y')

#set up empty variables 
abv = []
brewery_id = []
ibu_max = []
ibu_min = []
beer_name = []
brewery_inBusiness = []
beer_id = []
beer_style = []

for brewery in brewery_ID:
    url = (f'http://api.brewerydb.com/v2/brewery/{brewery}/beers/?key={api_key}&withBreweries=Y')
    response = requests.get(url).json()
    
    try:
        data = response["data"]
    except:
        pass
    
    for result in data:
        try: 
            abv.append(result['abv'])
        except (KeyError):
            abv.append("NaN")
        try: 
            brewery_id.append(result["breweries"][0]["id"])
        except (KeyError):
            brewery_id.append("NaN")
        try:
            ibu_max.append(result["style"]["ibuMax"])
        except (KeyError):
            ibu_max.append("NaN")
        try:
            ibu_min.append(result["style"]["ibuMin"])
        except (KeyError):
            ibu_min.append("NaN")
        try:
            beer_name.append(result["name"])
        except (KeyError):
            beer_name.append("NaN")
        try:
            brewery_inBusiness.append(result["breweries"][0]["isInBusiness"])
        except (KeyError):
            brewery_inBusiness.append("NaN")
        try:
            beer_id.append(result["id"])
        except (KeyError):
            beer_id.append("NaN")
        try:
            beer_style.append(result["style"]["name"])
        except (KeyError):
            beer_style.append("NaN")

    
 

```
