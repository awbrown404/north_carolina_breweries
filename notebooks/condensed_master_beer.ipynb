```python
import pandas as pd
```

```python
master_beer_df = pd.read_csv("../data/master_beer_df.csv")
```

```python
master_beer_df
```

```python
brewery = master_beer_df["breweries"]
beer_name = master_beer_df["beer_name"]
beer_style = master_beer_df["beer_style"]
master_style = master_beer_df["master_style"]
availability = master_beer_df["availability"]
abv = master_beer_df["abv"]
```

```python
columns = ["brewery", "beer_name", "beer_style", "availability", "aby", "master_style"]

master_beer_condensed = pd.DataFrame(columns=columns)

master_beer_condensed["brewery"] = brewery
master_beer_condensed["beer_name"] = beer_name
master_beer_condensed["beer_style"] = beer_style
master_beer_condensed["availability"] = availability
master_beer_condensed["abv"] = abv
master_beer_condensed["master_style"] = master_style

master_beer_condensed.to_csv("../data/master_beer_condensed.csv", index=False)
```
